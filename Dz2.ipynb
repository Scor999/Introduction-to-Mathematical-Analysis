1) Пусть $x_n = n^{(−1)^n}$. Доказать, что последовательность ${x_n}$:  
а) неограниченная;  
б) не является бесконечно большой.  


---
##  **1. Неограниченная**

Рассмотрим значения последовательности $x_n$ для нечётных и чётных $n$:
$x_{2k-1}=(2k-1)^{-1}$  
Эта последовательность стремится к нулю, когда
$k$ стремится к бесконечности. Это значит, что значения $x_{2k-1}$ никогда не будут большими и остаются ограниченными, но они не влияют на неограниченность всей последовательности.    
Для чётных $n$: $x_{2k}=(2k)^1=2k$  
Эта последовательность явно неограниченная, так как $2k$ стремится к бесконечности, когда $k$  стремится к бесконечности. Следовательно, последовательность $x_n$ неограниченная, так как она содержит неограниченную подпоследовательность $x_{2k}$

##  **2. Не является бесконечно большой**
Чтобы доказать, что последовательность $x_n$ не является бесконечно большой, нужно показать, что не все её элементы стремятся к бесконечности.  
Для нечётных $n$: $x_{2k-1}=(2k-1)^{-1}$    
Эта последовательность стремится к нулю, и, следовательно, значения остаются малыми для больших $k$.  
Поскольку подпоследовательность $x_{2k-1}$ стремится к нулю, это означает, что в последовательности $x_n$ всегда существуют значения, которые не стремятся к бесконечности.

Таким образом, хотя последовательность $x_{2k}$  стремится к бесконечности, последовательность $x_{2k-1}$  остаётся ограниченной. Следовательно, вся последовательность $x_n$ не является бесконечно большой.

  * Известно, что для иррационального $\alpha$ последовательность {$n\alpha$} плотно заполняет отрезок $[0,1]$.

  * Из-за непрерывности синуса и равномерного распределения углов {$n$ $mod$ $2\pi$}, значения $sin$$(n)$ будут распределены на отрезке $[-1,1]$.

  * Последовательность {$sin$ $n$} расходиться, если она не имеет предела.

Найти пределы:  
a) $\lim\limits_{n\to\infty}\frac{10n}{n^2+1}$    
б) $\lim\limits_{n\to\infty}\frac{n^2-n}{n-\sqrt n}$  
в) $\lim\limits_{n\to\infty}\frac{5\cdot3^n}{3^n-2}$

---
 ## Решение.
 * $\lim\limits_{n\to\infty}\frac{10n}{n^2+1}$
  1. $\frac{10n}{n^2+1}= \frac{10n}{n^2(1+\frac{1}{n^2})}=\frac{10}{n(1+\frac{1}{n^2})}$
  2. $\lim\limits_{n\to\infty}\frac{10}{n(1+\frac{1}{n^2})}=\lim\limits_{n\to\infty}\frac{10}{n}\cdot\frac{1}{1+\frac{1}{n^2}}=\lim\limits_{n\to\infty}\frac{10}{n}\cdot1=\lim\limits_{n\to\infty}\frac{10}{n}=0 $

## Ответ: **$\lim\limits_{n\to\infty}\frac{10n}{n^2+1}=0$**   
 * $\lim\limits_{n\to\infty}\frac{n^2-n}{n-\sqrt n}$  
  $\lim\limits_{n\to\infty} \frac{n^2-n}{n-\sqrt{n}} = \frac{\infty}{\infty} = \lim\limits_{n\to\infty} \frac{\infty^2-\infty}{\infty-\sqrt{\infty}} = \lim\limits_{n\to\infty} \frac{\infty-\infty}{\infty-\infty} = \infty$
## Ответ:  **$\lim\limits_{n\to\infty}\frac{n^2-n}{n-\sqrt n}= \infty$**
 * $\lim\limits_{n\to\infty}\frac{5\cdot3^n}{3^n-2}$
  1. $\frac{5\cdot3^n}{3^n-2}=\frac{5\cdot3^n /3^n}{(3^n-2)/3}=\frac{5}{1-\frac{2}{3^n}}$     
  При $n\to\infty$, $\frac{2}{\frac{2}{3^n}}\to0$ потому что $3^n$ растет экспоненциально быстрее, чем линейный рост числителя. Поэтому выражение становится:   
  $$\lim\limits_{n\to\infty}\frac{5}{1-0}=\frac{5}{1}=5$$
## Ответ: **$\lim\limits_{n\to\infty}\frac{5\cdot3^n}{3^n-2}=5$**

4) Найти предел $\lim\limits_{n\to\infty} (\sqrt{n^2+n} - n)$

---

## Решение.
Умножим и разделим на сопряженное выражение $\sqrt{n^2+n} + n$    

$$\frac{(\sqrt{n^2+n} - n)(\sqrt{n^2+n} + n)}{\sqrt{n^2+n} + n} = \frac{(\sqrt{n^2+n})^2 + n\cdot \sqrt{n^2+n} -n\cdot \sqrt{n^2+n} - n^2}{\sqrt{n^2+n} + n} = \frac{n}{\sqrt{n^2+n} + n}$$


разделим числитель и знаменатель на $n$   

$$\frac{\frac{n}{n}}{\sqrt{n^2+n}/n + 1} = \frac{1}{\sqrt{1+\frac{1}{n}}+1}$$

исходя из того что $\frac{1}{n} \to 0$ при $n \to \infty$ получим

$$\lim\limits_{n\to\infty} (\sqrt{n^2+n} - n)=\frac{1}{\sqrt{1+0}+1} = \frac{1}{2}$$

## Ответ: **$\lim\limits_{n\to\infty} (\sqrt{n^2+n} - n)=\frac{1}{2}$**

5) Вычислить $\lim\limits_{n\to\infty} \frac{\sqrt{n}\cdot \cos{n}}{n+1}$

---

## Решение.

Числитель $\sqrt n\cdot \cos n$ одержит осциллирующую функцию $\cos n$, значения которой лежат в интервале $[-1,1]$. Таким образом,числитель будет колебаться между $-\sqrt n$ и $\sqrt n$.   
Знаменатель $n+1$ растет линейно и стремится к бесконечности при $n\to\infty$.
Обсолютное выражение:
$$|\frac {\sqrt n\cdot \cos n}{n+1}|\leq\frac{\sqrt n}{n+1}$$
Предел правой части:
$$$\lim\limits_{n\to\infty}\frac {\sqrt n}{n+1}= \frac {\sqrt n /n}{(n+1)/n}=\frac{1/\sqrt n}{1+1/n}$$
При $n\to \infty$,$\frac{1}{\sqrt n}\to 0$ и $\frac {1} {n} \to 0$ :
$$|\frac {\sqrt n\cdot \cos n}{n+1}| \to 0$$


Следовательно так как абсолютное значение стремиться к нолю, исходный предел также равен нулю.
## Ответ:
$$\lim\limits_{n\to\infty} \frac{\sqrt{n}\cdot \cos{n}}{n+1} \to 0$$
