1. Представьте в виде несократимой рациональной дроби:    
а) 0.(216);   
б) 1.0(01).

Рассмотрим каждое из чисел по отдельности и представим их в виде несократимой рациональной дроби.

##а) 0.(216):

###Решение

----

Обозначим
\begin{equation*}
x = 0.(216)
\end{equation*}
Умножим равенство на такое число чтобы период оказался в целой части т.е. на 1000
\begin{equation*}
1000x = 216.(216)  
\end{equation*}
Вычтем полученные равенства
\begin{equation*}
1000x - x= 216.(216) - 0.(216)
\end{equation*}
999x = 216
\end{equation*}
\begin{equation*}  
x = \frac{216}{999}
\end{equation*}
Теперь сократим дробь:

НОД чисел 216 и 999 равен 27.
Разделим числитель и знаменатель на 27:   .
\begin{equation*}
x =\frac{8}{37}
\end{equation*}
\begin{equation*}  
0.(216) =\frac{8}{37}
\end{equation*}

---

##Ответ:
\begin{equation*}
0.(216)=\frac{8}{37}
\end{equation*}

-----


##б) 1.0(01)
###Решение

---


Обозначим:
\begin{equation*}    
y = 1.0(01)
\end{equation*}
Умножим равенство на такое число чтобы период оказался в целой части т.е. на 1000  
\begin{equation*}   
 1000y = 1001.(01)
 \end{equation*}  
Теперь вычтем исходное уравнение из нового:
\begin{equation*}
1000y-y=1001.001001…-1.001001…       
\end{equation*}
\begin{equation*}
999y=1000
\end{equation*}
Разделим обе стороны на 999:  
\begin{equation*}
y=\frac{1000}{999}
\end{equation*}
Таким  образом,
\begin{equation*}
1.0.(01)=\frac{1000}{999}
\end{equation*}

Так как НОД равен 1, дробь уже является несократимой.

---

##Ответ:
\begin{equation*}
1.0.(01)=\frac{1000}{999}
\end{equation*}



2. Проверьте любым способом, являются ли данные логические формулы
тавтологией:    
а)
\begin{equation*}
 (A ∨ B) → (B ∨ \overline A)
 \end{equation*}
б)
\begin{equation*}
A → (A ∨ (\overline B ∧ A))
\end{equation*}

Составим таблицы истинности для данных формул:
а) ($A$∨$B$)→($B$∨$\overline A$)

|$A$|$B$|$A$∨$B$|$A$|$B$∨$\overline A$| ($A$∨$B$)→($B$∨$\overline A$)|

|0  |1  |1    |1            |1              |1                      |
|1  |0  |1    |0            |0              |0                      |
|1  |1  |1    |0            |1              |1                      |

Формула ($A$∨$B$)→($B$∨$\overline A$)  не является тавтологией, так как принимает значение "ложь" при $A=1$ и $B=0$

б) $A → (A ∨ (\overline B ∧ A))$

|$A$|$B$|$\overline B$|$\overline B$∧$A$|$A$∨($\overline B$∧$A$)|$A$→($A$∨($\overline B$∧$A$))|

|0   |0   |1   |0   |0   |1   |
|0   |1   |0   |0   |0   |1   |

Формула $A$→($A$∨($\overline B$∧$A$)) является тавтологией, так как принимает значение "истина" при любых значениях переменных $A$ и $B$
